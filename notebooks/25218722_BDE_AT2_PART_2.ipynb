### 1. Year-month summary with total trips, Most trip day of week, Most trip hour of day, average passengers, average amount per trip and average amount per passenger.

```python
%sql
-- 1: Year-month summary
WITH base AS (
  SELECT
    date_format(pickup_datetime, 'yyyy-MM') AS year_month,
    date_format(pickup_datetime, 'EEEE') AS day_week,
    hour(pickup_datetime) AS hour_of_day,
    passenger_count,
    total_amount
  FROM final_taxi_data
),
agg_ym AS (
  SELECT
    year_month,
    COUNT(*) AS total_trips,
    ROUND(AVG(passenger_count), 2) AS avg_passengers,
    ROUND(AVG(total_amount), 2) AS avg_amount_per_trip,
    ROUND(AVG(total_amount / passenger_count), 2) AS avg_amount_per_passenger
  FROM base
  GROUP BY year_month
),
top_day_ranked AS (
  SELECT year_month, day_week AS top_day_of_week
  FROM (
    SELECT year_month, day_week, COUNT(*) AS cnt,
           ROW_NUMBER() OVER (PARTITION BY year_month ORDER BY COUNT(*) DESC, day_week) AS rn
    FROM base
    GROUP BY year_month, day_week
  ) WHERE rn = 1
),
top_hour_ranked AS (
  SELECT year_month, hour_of_day AS top_hour_of_day
  FROM (
    SELECT year_month, hour_of_day, COUNT(*) AS cnt,
           ROW_NUMBER() OVER (PARTITION BY year_month ORDER BY COUNT(*) DESC, hour_of_day) AS rn
    FROM base
    GROUP BY year_month, hour_of_day
  ) WHERE rn = 1
)
SELECT
  a.year_month,
  a.total_trips,
  tdr.top_day_of_week,
  thr.top_hour_of_day,
  a.avg_passengers,
  a.avg_amount_per_trip,
  a.avg_amount_per_passenger
FROM agg_ym a
LEFT JOIN top_day_ranked tdr ON a.year_month = tdr.year_month
LEFT JOIN top_hour_ranked thr ON a.year_month = thr.year_month
ORDER BY a.year_month;
```

### 2. For each taxi colour calculate average, median, minimum and maximum of trip duration, time and speed.

```python
%sql
-- 2. Per taxi_colour summary
SELECT
    taxi_colour,
    -- Trip duration in minutes
    ROUND(AVG(trip_time * 60), 2) AS avg_duration_m,
    ROUND(percentile_approx(trip_time * 60, 0.5), 2) AS median_duration_m,
    ROUND(MIN(trip_time * 60), 2) AS min_duration_m,
    ROUND(MAX(trip_time * 60), 2) AS max_duration_m,

    -- Trip distance in km
    ROUND(AVG(trip_distance * 1.60934), 2) AS avg_distance_km,
    ROUND(percentile_approx(trip_distance * 1.60934, 0.5), 2) AS median_distance_km,
    ROUND(MIN(trip_distance * 1.60934), 2) AS min_distance_km,
    ROUND(MAX(trip_distance * 1.60934), 2) AS max_distance_km,

    -- Speed in km/hr
    ROUND(AVG(speed_mph * 1.60934), 2) AS avg_speed_kmh,
    ROUND(percentile_approx(speed_mph * 1.60934, 0.5), 2) AS median_speed_kmh,
    ROUND(MIN(speed_mph * 1.60934), 2) AS min_speed_kmh,
    ROUND(MAX(speed_mph * 1.60934), 2) AS max_speed_kmh

FROM final_taxi_data
GROUP BY taxi_colour;
```

### 3. Trips summary by taxi colour, borough pair, month, day of week, and hour

```python
%sql
-- 3. Trips summary by taxi colour, borough pair, month, day of week, and hour
SELECT
    taxi_colour,
    pickup_borough,
    dropoff_borough,
    MONTH(pickup_datetime) AS month,
    DATE_FORMAT(pickup_datetime, 'EEEE') AS day_of_week,
    HOUR(pickup_datetime) AS hour,

    COUNT(*) AS total_trips,
    ROUND(AVG(trip_distance * 1.60934), 2) AS avg_distance_km,
    ROUND(AVG(total_amount), 2) AS avg_amount_per_trip,
    ROUND(SUM(total_amount), 2) AS total_amount_paid

FROM final_taxi_data
GROUP BY
    taxi_colour,
    pickup_borough,
    dropoff_borough,
    MONTH(pickup_datetime),
    DATE_FORMAT(pickup_datetime, 'EEEE'),
    dayofweek(pickup_datetime),     -- number format of week 1=Sunday ... 7=Saturday
    HOUR(pickup_datetime)
ORDER BY
    taxi_colour,
    month,
    dayofweek(pickup_datetime),   -- ensures proper week order
    hour;
```

### 4.  Share of total revenue contributed by the Top 10 pickupdropoff borough pairs in 2024.

```python
%sql
-- 4. Top 10 pickupdropoff borough pairs by total revenue in 2024
WITH revenue_2024 AS (
    SELECT
        pickup_borough,
        dropoff_borough,
        ROUND(SUM(total_amount),2) AS total_revenue
    FROM final_taxi_data
    WHERE YEAR(pickup_datetime) = 2024
    GROUP BY pickup_borough, dropoff_borough
),
total_rev AS (
    SELECT ROUND(SUM(total_revenue),2) AS grand_total
    FROM revenue_2024
)
SELECT
    r.pickup_borough,
    r.dropoff_borough,
    ROUND((r.total_revenue / t.grand_total) * 100, 2) AS revenue_share_percent
FROM revenue_2024 r
CROSS JOIN total_rev t
ORDER BY r.total_revenue DESC
LIMIT 10;
```

### 5. Percentage of trips where drivers received tips

```python
%sql
-- 5. Percentage of trips where drivers received tips
SELECT 
    ROUND( (SUM(CASE WHEN tip_amount > 0 THEN 1 ELSE 0 END) * 100.0) / COUNT(*), 2 ) 
    AS pct_trips_with_tips
FROM final_taxi_data;
```

-----
### 6. Percentage of trips where drivers received tips Atleast 15.

```python
%sql
-- 6. Percentage of tipped trips where tip >= 15
SELECT ROUND(
        (SUM(CASE WHEN tip_amount >= 15 THEN 1 ELSE 0 END) * 100.0) 
        / COUNT(*), 2
    ) AS pct_trips_with_15plus
FROM final_taxi_data
WHERE tip_amount > 0;
```

### 7. Trip duration bins with average speed and distance per dollar

```python
%sql
-- 7. Trip duration bins with average speed and distance per dollar
SELECT
    CASE 
        WHEN trip_time * 60 < 5 THEN 'Under 5 Mins'
        WHEN trip_time * 60 BETWEEN 5 AND 10 THEN '510 Mins'
        WHEN trip_time * 60 BETWEEN 10 AND 20 THEN '1020 Mins'
        WHEN trip_time * 60 BETWEEN 20 AND 30 THEN '2030 Mins'
        WHEN trip_time * 60 BETWEEN 30 AND 60 THEN '3060 Mins'
        ELSE 'At least 60 Mins'
    END AS duration_bin,

    ROUND(AVG(speed_mph * 1.60934), 2) AS avg_speed_kmh,
    ROUND(AVG((trip_distance * 1.60934) / NULLIF(total_amount, 0)), 2) AS avg_distance_per_dollar_km
FROM final_taxi_data
GROUP BY duration_bin
ORDER BY 
    CASE duration_bin
        WHEN 'Under 5 Mins' THEN 1
        WHEN '510 Mins' THEN 2
        WHEN '1020 Mins' THEN 3
        WHEN '2030 Mins' THEN 4
        WHEN '3060 Mins' THEN 5
        ELSE 6
    END;
```

### 8. Duration Bin to target to Maximise Driver Income

```python
%sql
-- 8. Revenue efficiency by trip duration bins
SELECT
    CASE 
        WHEN trip_time * 60 < 5 THEN 'Under 5 Mins'
        WHEN trip_time * 60 BETWEEN 5 AND 10 THEN '510 Mins'
        WHEN trip_time * 60 BETWEEN 10 AND 20 THEN '1020 Mins'
        WHEN trip_time * 60 BETWEEN 20 AND 30 THEN '2030 Mins'
        WHEN trip_time * 60 BETWEEN 30 AND 60 THEN '3060 Mins'
        ELSE 'At least 60 Mins'
    END AS duration_bin,

    COUNT(*) AS total_trips,
    ROUND(AVG(total_amount), 2) AS avg_amount_per_trip,
    ROUND(AVG(total_amount / NULLIF(trip_time * 60, 0)), 2) AS avg_revenue_per_min
FROM final_taxi_data
GROUP BY duration_bin
ORDER BY avg_revenue_per_min DESC;
```
